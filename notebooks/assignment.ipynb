### Task 1: Time Series Setup

    Create a dummy dataframe (or load one) with an index of daily dates for the year 2023.
    Generate random "Sales" data for each day.
    Resample this daily data to find the Total (Sum) Sales per Month.


```python
# create a dummy dataframe with an index of daily dates for the year 2023
import pandas as pd

df = pd.DataFrame(index=pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'))
df
```

```python
# generate a random number 'Sales' for each day
import numpy as np

df['Sales'] = np.random.randint(100, 1000, len(df))
df
```

```python
# resample to a monthly frequency
df_monthly = df.resample('M').sum()
df_monthly
```

### Task 2: The "Wide" Report

    Create a small dataframe with columns: ['Date', 'Region', 'Sales']. Populate it with data for 3 regions (North, South, East) over 5 different dates.
    Use pivot() to reshape this data so that:
        Date is the Index.
        Region names are the Columns (North, South, East).
        Sales are the Values.

```python
# Create data for 3 regions over 5 dates
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-01',
             '2023-01-02', '2023-01-02', '2023-01-02',
             '2023-01-03', '2023-01-03', '2023-01-03',
             '2023-01-04', '2023-01-04', '2023-01-04',
             '2023-01-05', '2023-01-05', '2023-01-05'],
    'Region': ['North', 'South', 'East'] * 5,
    'Sales': np.random.randint(200, 500, size=15)
}

# Create the dataframe
df_wide = pd.DataFrame(data)
print("Original DataFrame:")
print(df_wide)

# Use pivot() to reshape the data
pivot_df = df_wide.pivot(index='Date', columns='Region', values='Sales')
print("\nPivot Table (Wide Format):")
print(pivot_df)
```

### Task 3: The Executive Summary

    Using the tips dataset (or similar), create a Pivot Table that shows:
        Rows: Day of the Week.
        Columns: Time (Lunch vs. Dinner).
        Values: Total Bill.
        Aggregation: Calculate the Average (mean) bill size.
        Margins: Include the "All" row/column totals.

```python
# load dataset from seaborn
import seaborn as sns
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
# Rename day categories (Thur->Thursday, etc.)
tips['day'] = tips['day'].astype('category')
tips['day'] = tips['day'].cat.rename_categories({'Thur': 'Thursday', 'Fri': 'Friday', 'Sat': 'Saturday', 'Sun': 'Sunday'})
# Don't convert to codes - keep the day names as strings!

# The original dataset already has 'Lunch' and 'Dinner'
tips['time'] = tips['time'].astype('category')
# Don't convert to codes - keep "Lunch" and "Dinner" as strings!

tips['total_bill'] = tips['total_bill'].astype('int')

print("Cleaned DataFrame:")
print(tips.head())
print("\nUnique days:", tips['day'].unique())
print("Unique times:", tips['time'].unique())
print("Unique total bills:", tips['total_bill'].unique())
```

```python
# Create pivot table for Task 3
# rows = day, columns = time
# time = 'Lunch' and 'Dinner'
# values = total_bill
# aggfunc = mean
# margins = True
pivot_table = pd.pivot_table(
    data=tips, 
    index='day', 
    columns='time', 
    values='total_bill', 
    aggfunc='mean', 
    margins=True)
print(pivot_table)
```

```python
# Apply formatting to the result
formatted_table = pivot_table.applymap('${:,.2f}'.format)
print("\nformatted_table\n", formatted_table)
```
